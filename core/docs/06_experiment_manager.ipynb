# Experiment Manager

This notebook shows how to use the `Manager` class to run and compare multiple active learning experiments. The `Manager` loads experiment definitions from a YAML config file and runs them in sequence or in parallel.

```python
import sys
from pathlib import Path
import matplotlib.pyplot as plt

PROJECT_ROOT = Path().absolute().parent.parent
sys.path.insert(0, str(PROJECT_ROOT))

from core.active_learner import Manager
```

## 1. Load experiments from config

The config file (`core/config.yml`) defines each experiment with its sampling strategy, paths, and hyperparameters.

```python
config_path = PROJECT_ROOT / "core" / "config.yml"
manager = Manager(config_path, base_dir=PROJECT_ROOT, verbose=False)

print(f"Loaded {len(manager.experiments)} experiments:")
for name in manager.experiment_names:
    print(f"  - {name}")
```

## 2. Run active learning cycles

Each call to `manager.run()` executes one cycle (sample, add, train) across all experiments.

```python
n_cycles = 10

for cycle in range(n_cycles):
    results = manager.run(
        n_samples=10,
        epochs=10,
        batch_size=8,
        parallel=False
    )

    if cycle % 3 == 0 or cycle == n_cycles - 1:
        print(f"\nCycle {cycle + 1}/{n_cycles}")
        for exp_name, metrics in results.items():
            print(f"  {exp_name}: accuracy={metrics['accuracy']:.4f}, labelled={metrics['n_labeled']}")
```

## 3. Compare training histories

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

for learner, name in zip(manager.experiments, manager.experiment_names):
    accuracies = [h['accuracy'] for h in learner.training_history]
    losses = [h['loss'] for h in learner.training_history]
    n_labeled = [h['n_labeled'] for h in learner.training_history]

    axes[0].plot(n_labeled, accuracies, marker='o', label=name)
    axes[1].plot(n_labeled, losses, marker='o', label=name)

axes[0].set_xlabel('Labelled samples')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Accuracy comparison')
axes[0].legend()
axes[0].grid(True)

axes[1].set_xlabel('Labelled samples')
axes[1].set_ylabel('Loss')
axes[1].set_title('Loss comparison')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## 4. Save results

```python
manager.save(output_dir=PROJECT_ROOT / "results" / "experiment_comparison")
print("Results saved.")
```

## 5. Adding experiments at runtime

You can also add new experiments programmatically without editing the config file.

```python
new_config = {
    'embeddings_dir': PROJECT_ROOT / 'ESC10_BASEAL' / 'embeddings' / 'birdnet',
    'annotations_path': PROJECT_ROOT / 'ESC10_BASEAL' / 'labels.csv',
    'sampling_strategy': 'random',
    'learning_rate': 0.0005,
    'device': 'cpu'
}

manager.add(new_config, name='random_low_lr')
print(f"Total experiments: {len(manager.experiments)}")
```

## Summary

| Method | What it does |
|--------|-------------|
| `Manager(config_path)` | Loads experiments from YAML config |
| `manager.run(n_samples, epochs)` | Runs one AL cycle across all experiments |
| `manager.save(output_dir)` | Saves results to JSON files |
| `manager.add(config, name)` | Adds a new experiment at runtime |
| `manager.get_summary()` | Returns current status of all experiments |
