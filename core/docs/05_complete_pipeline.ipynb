# Complete Pipeline

This notebook demonstrates the full active learning pipeline using the `ActiveLearner` class directly. If you just want to run experiments from the config file, see notebook 01 (Getting Started) or notebook 06 (Experiment Manager) instead.

We will:
1. Initialise an `ActiveLearner` with the ESC10 dataset
2. Run several active learning cycles
3. Plot the training progress

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

PROJECT_ROOT = Path().absolute().parent.parent
sys.path.insert(0, str(PROJECT_ROOT))

from core.active_learner import ActiveLearner
```

## Step 1: Initialise

Point the learner at the embeddings directory and labels CSV. The learner loads the data, creates the model, and sets up the labelled/unlabelled index sets.

```python
EMBEDDINGS_DIR = PROJECT_ROOT / "ESC10_BASEAL" / "embeddings" / "birdnet"
LABELS_PATH = PROJECT_ROOT / "ESC10_BASEAL" / "labels.csv"

learner = ActiveLearner(
    embeddings_dir=EMBEDDINGS_DIR,
    annotations_path=LABELS_PATH,
    model_name="birdnet",
    dataset_name="ESC10",
    hidden_dim=None,
    learning_rate=0.001,
    sampling_strategy="margin",
    device="cpu",
    verbose=False
)

state = learner.get_state()
print(f"Total samples: {state['n_labeled'] + state['n_unlabeled']}")
print(f"Labelled: {state['n_labeled']}")
print(f"Unlabelled: {state['n_unlabeled']}")
print(f"Classes: {state['num_classes']}")
```

## Step 2: Run active learning cycles

Each cycle samples new points, adds them to the labelled set, and trains the model.

```python
n_cycles = 10
n_samples_per_cycle = 10
epochs_per_cycle = 10

history = []

for cycle in range(n_cycles):
    # Sample, add, train
    selected = learner.sample(n_samples=n_samples_per_cycle)
    learner.add_samples(selected)
    metrics = learner.train_step(epochs=epochs_per_cycle, batch_size=8)
    history.append(metrics)

    if cycle % 3 == 0 or cycle == n_cycles - 1:
        print(f"Cycle {cycle+1}: accuracy={metrics['accuracy']:.4f}, loss={metrics['loss']:.4f}, labelled={metrics['n_labeled']}")
```

## Step 3: Plot results

```python
losses = [h['loss'] for h in history]
accuracies = [h['accuracy'] for h in history]
n_labeled = [h['n_labeled'] for h in history]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(n_labeled, accuracies, marker='o')
axes[0].set_xlabel('Labelled samples')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Accuracy')
axes[0].grid(True)

axes[1].plot(n_labeled, losses, marker='o', color='orange')
axes[1].set_xlabel('Labelled samples')
axes[1].set_ylabel('Loss')
axes[1].set_title('Loss')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## Summary

The `ActiveLearner` class provides three key methods:

| Method | What it does |
|--------|-------------|
| `sample(n_samples)` | Selects the most informative unlabelled samples |
| `add_samples(indices)` | Moves selected samples to the labelled set |
| `train_step(epochs, batch_size)` | Trains the model and returns metrics |
